## Check Environment - [ðŸ”™](#TOC)

```python
# Get python version
# Check Python and Jupyter versions

import sys
import platform

# Python version
print("Python Version:")
print(sys.version)
print(f"Python {sys.version_info.major}.{sys.version_info.minor}.{sys.version_info.micro}")

# Platform info
print(f"\nPlatform: {platform.platform()}")

# Check if running in Jupyter
try:
    get_ipython()
    print("\nEnvironment: Jupyter Notebook/Lab")
except NameError:
    print("\nEnvironment: Regular Python script")

# Check Jupyter-related packages
print("\n" + "="*50)
print("JUPYTER PACKAGE VERSIONS")
print("="*50)

# IPython
try:
    import IPython
    print(f"IPython: {IPython.__version__}")
except ImportError:
    print("IPython: Not installed")

# Jupyter Notebook
try:
    import notebook
    print(f"Jupyter Notebook: {notebook.__version__}")
except ImportError:
    print("Jupyter Notebook: Not installed")

# JupyterLab
try:
    import jupyterlab
    print(f"JupyterLab: {jupyterlab.__version__}")
except ImportError:
    print("JupyterLab: Not installed")

# Jupyter Core
try:
    import jupyter_core
    print(f"Jupyter Core: {jupyter_core.__version__}")
except ImportError:
    print("Jupyter Core: Not installed")

# Using subprocess to check with pip (alternative method)
import subprocess
import sys

print("\n" + "="*50)
print("CHECKING WITH PIP")
print("="*50)

packages = ['jupyter', 'jupyterlab', 'notebook', 'ipython']

for package in packages:
    try:
        result = subprocess.run([sys.executable, '-m', 'pip', 'show', package], 
                              capture_output=True, text=True, timeout=5)
        if result.returncode == 0:
            for line in result.stdout.split('\n'):
                if line.startswith('Version:'):
                    version = line.split(': ')[1]
                    print(f"{package}: {version}")
                    break
        else:
            print(f"{package}: Not found")
    except Exception as e:
        print(f"{package}: Error - {e}")
```
