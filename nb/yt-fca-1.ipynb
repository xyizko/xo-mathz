````python
# ---------------------------------------
# Solving for multiple unknowns
# ---------------------------------------

from sympy import symbols, Eq, solveset, sympify, solve

def two_vars():

    # Defining two variables for solving
    x, y = symbols('x y') 

    # Equations Array
    equations = [
        Eq(2*x + 10, x),
        Eq(2*x + 10, y),
    ]

    try:
        sol1 = solveset(equations[0],x) # 2x + 10 = x
        sol2 = solveset(equations[1],x) # 2x + 10 = y 
        sol3 = solveset(equations[0],y) 
        sol4 = solveset(equations[1],y)

        rprint(f"Soln for x = {equations[0]} = {sol1}")          
        rprint(f"Soln for x = {equations[1]} = {sol2}")          
        rprint(f"Soln for y = {equations[0]} = {sol3}")          
        rprint(f"Soln for y = {equations[1]} = {sol4}")          
        

    except Exception as e:
        console.print(f"[red]Error: {e}[/red]")

# Execute Function
two_vars()
````

````python
# Solving both X and Y 

def solve_system():
    x, y = symbols('x y')
    
    # System of equations
    equations = [
        Eq(2*x + 10, x),    # 2x + 10 = x
        Eq(2*x + 10, y)     # 2x + 10 = y
    ]
    
    # Solve system
    solution = solve(equations, [x, y])
    rprint(f"System solution: {solution}")  # {x: -10, y: -10}

solve_system()
````

````python
# Fixed Code 

def string_frac(frac_str):
    """Convert string fraction to float"""
    try:
        if '/' in frac_str:
            num, den = frac_str.split('/')
            return float(num) / float(den)
        else:
            return float(frac_str)
    except:
        return None

def one_step_mult():
    # Generate random numbers
    a = random.randint(1, 11)
    b = random.randint(2, 24)

    # Display the equation: a * x = b, solve for x
    rprint(f"{a}x = {b}")
    rprint("[yellow]Solve for x:[/yellow]")
    
    ans_in = input("x = ")
    
    try:
        # Calculate correct answer
        correct_answer = b / a
        
        # Test user answer
        user_answer = string_frac(ans_in)
        
        if user_answer is not None and abs(user_answer - correct_answer) < 1e-10:
            rprint("[green]Correcto![/green] ðŸŽ‰")
        else:
            rprint(f"[red]Incorrect![/red] The answer is [cyan]{correct_answer}[/cyan]")
            if correct_answer == int(correct_answer):
                rprint(f"That's [cyan]{int(correct_answer)}[/cyan]")
            else:
                rprint(f"As a fraction: [cyan]{b}/{a}[/cyan]")
            
    except Exception as e:
        console.print(f"[red]Error: {e}[/red]")

# Execute 
one_step_mult()
````

## Fractions and Decimals

````python
#--------------------------
# Input Casting for Fractions 
#--------------------------

def get_numerator():

    # get input
    digits = Prompt.ask("Enter decimal number places to correct")  #âž¡ï¸  0.125

    # calculate number of decomal places 
    exponent = int(len(digits))-1  #âž¡ï¸ 4-1 = 3

    # Cast to float for precision
    n = float(digits)  #âž¡ï¸  0.125

    # Get numerator 
    numerator = int (n * 10**exponent) #âž¡ï¸  0.125 * 10^3 - 0.125 * 1000 = 125 

    # use exponent to get denominator
    denominator = 10**exponent #âž¡ï¸  10**3 = 1000
    
    # percent is the first two decimal places 
    percent = n * 100  #âž¡ï¸  0.125 * 100 = 12.5

    # Print it out 
    rprint("Decimal : ",n) #âž¡ï¸  0.125
    rprint("Fraction : ", numerator, "/", denominator) #âž¡ï¸  125 / 1000
    rprint("Percent : ", percent, "%") #âž¡ï¸  12.5

get_numerator()
````

````python
from sympy import symbols, Eq, solveset, sympify, solve, factor, latex 
from rich import print as rprint

def dividerz(text):
    rprint(f"\n{'='*20} {text} {'='*20}\n")

def fase1():
    # Identify all variables
    a, b, c, d, x = symbols('a b c d x')

    # Equation
    equn1 = "a*x**2 + b*x + c" 
    eqn = Eq(0, sympify(equn1))

    # Variables to solve for 
    variables = [x, a, b, c]  

    # Solve for each variable
    rprint("[yellow]Solving quadratic equation: a*xÂ² + b*x + c = 0[/yellow]")
    
    for var in variables:
        try:
            dividerz(f"Solving for {var}")
            rprint(f"Equation: {eqn}")
            sol = solveset(eqn, var)
            rprint(f"Solved {var} = {sol}")
            rprint(f"Factoring {sol}")
            to_latex = sol
            rprint(latex(to_latex))
        except Exception as e:
            rprint(f"[red]Error solving for {var}: {e}[/red]")

# -- Execute ---
fase1()
````
