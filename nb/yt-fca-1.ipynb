### Solving the following 

$$
\frac{1}{2} = \frac{x}{16}
$$
$$
\text{Solve for x}
$$

````python
# PRO1 

def func1():

    # Declaring the variables 
    n1 = 1
    d1 = 2
    n2 = 0
    d2 = 16

    # Conditional Checks

    if n2 == 0:
        answer = d2 * n1 / d1
        rprint("n2 = ", answer)

    if d2 == 0:
        answer = n2 * d1 / n1
        rprint("d2 = ", answer)
````

> Solving for two unknowns

equation that is being solved 

$$
\begin{gather*}
\text{Solve for x, y}\\
2x + 10 = y\\
\end{gather*}
$$

````python
# ---------------------------------------
# Solving for multiple unknowns
# ---------------------------------------

from sympy import symbols, Eq, solveset, sympify, solve

def two_vars():

    # Defining two variables for solving
    x, y = symbols('x y') 

    # Equations Array
    equations = [
        Eq(2*x + 10, x),
        Eq(2*x + 10, y),
    ]

    try:
        sol1 = solveset(equations[0],x) # 2x + 10 = x
        sol2 = solveset(equations[1],x) # 2x + 10 = y 
        sol3 = solveset(equations[0],y) 
        sol4 = solveset(equations[1],y)

        rprint(f"Soln for x = {equations[0]} = {sol1}")          
        rprint(f"Soln for x = {equations[1]} = {sol2}")          
        rprint(f"Soln for y = {equations[0]} = {sol3}")          
        rprint(f"Soln for y = {equations[1]} = {sol4}")          
        

    except Exception as e:
        console.print(f"[red]Error: {e}[/red]")

# Execute Function
two_vars()
````

````python
# Solving both X and Y 

def solve_system():
    x, y = symbols('x y')
    
    # System of equations
    equations = [
        Eq(2*x + 10, x),    # 2x + 10 = x
        Eq(2*x + 10, y)     # 2x + 10 = y
    ]
    
    # Solve system
    solution = solve(equations, [x, y])
    rprint(f"System solution: {solution}")  # {x: -10, y: -10}

solve_system()
````

## Factoring Equations

````python
#----------------------------------------------------------
# Factoring TEquations
#----------------------------------------------------------

from sympy import symbols, Eq, solveset, sympify, solve, factor, latex 

def factoring_equation1() :

    # Define symbols for ocmputation
    x,y = symbols('x y')

    # Equations 

    eqz = [
        Eq(2*x + 10*y + 4,0), # 2x + 10y + 4
        Eq(x**2-4,0)
    ]

    try:
        for e in eqz:
            rprint(f"{e}")
            
            dividerz("Solve for X ", "white")
            rprint(f"{solve(e,x)}")

            dividerz("Solve for Y ", "white")
            rprint(f"{solve(e,y)}")

            dividerz("Factor x ", "white")
            rprint(f"{factor(e)}") # Factorization

            # Testing ltex print 
            dividerz("Factor Latex Print ", "white")
            to_latex = factor(e)
            rprint(latex(to_latex))

            dividerz("end ", "red", "red")

            
            

    except Exception as e:
        console.print(f"[red]Error: {e}[/red]")

# --- Execute --- 
factoring_equation1()
````
