```python
# Indexing access 

def ind1():

    # Setting up array and series with mpy 
    is1 = np.random.randint(0,10, 10)
    is1d = pd.Series(is1)

    # Adding Index 
    is1d_index = [
    "Apple", "Banana", "Mango", "Orange", "Pineapple",
    "Grapes", "Strawberry", "Watermelon", "Blueberry", "Papaya"] 

    is1d.index = is1d_index

    # Printing
    rprint(
f"""
is1 = \n {is1} \n

is1d = \n {is1d} \n

--- Index based Access -

Location 0 = {is1d[0]}
Range = \n {is1d[0:5]}

--- Iloc access ---
At position 9 = {is1d.iloc[9]}
Multiple Positions = \n{is1d.iloc[[3,1,5]]}

--- Loc --- 
Calling my label = \n {is1d.loc["Apple": "Grapes"]} \n
Multi Element Access = \n {is1d.loc[["Apple", "Grapes"]]} \n
""")

ind1()
```

```python
# Dictionary Acess 

def dik1():

    boxa("Dictionary and Pandas")

    # Dictionary 
    fr_ca = {
    "Apple": 52.0,
    "Banana": 89.0,
    "Mango": 60.0,
    "Orange": 47.0,
    "Pineapple": 50.0,
    "Grapes": 69.0,
    "Strawberry": 33.0,
    "Watermelon": 30.0,
    "Blueberry": 57.0,
    "Papaya": 43.0
    }

    # Converting the dictionary to series 
    fr_ca_s = pd.Series(fr_ca)
    fr_ca_s.name = "Kalories" 

    # --- Printing ---
    
    pprint(fr_ca_s)
    rprint(
f"""
Fruits and Caliries Pandas Series = \n {fr_ca_s} \n

--- Conditional Acess --- 

Calories>50 =
{fr_ca_s > 50}

Actual Data with condition >50 =
{fr_ca_s[fr_ca_s > 50]}

--- Logical Operators series --- 

Find Fruits Caloreis > 40 & < 80 = 
{(fr_ca_s > 40) & (fr_ca_s < 80)}

Actual data of Above Condition
{fr_ca_s[(fr_ca_s >= 40) & (fr_ca_s <= 80)]}

Or Operation - Yeild all values because of ur range
{fr_ca_s[(fr_ca_s > 60) | (fr_ca_s < 80)]}

Not Operation - yeild all value not greater than 50 (meaning less than)
{fr_ca_s[~(fr_ca_s > 50)]}


 
""")

dik1()
```

## Joins and Merges of Datasets

```python
#  Data frames


fruits_data = {
    "Fruit": ["Apple", "Banana", "Mango", "Orange", "Pineapple", "Grapes", "Strawberry", "Watermelon", "Blueberry", "Papaya"],
    "Color": ["Red", "Yellow", "Orange", "Orange", "Brown", "Green", "Red", "Green", "Blue", "Orange"],
    "Avg_Weight_g": [182, 118, 200, 130, 900, 80, 12, 5000, 80, 500]
}

fruits_df = pd.DataFrame(fruits_data)
```

```python
fruits_df
```

```python
veges_data = {
    "Vegetable": ["Carrot", "Broccoli", "Spinach", "Tomato", "Potato", "Onion", "Lettuce", "Cabbage", "Peas", "Pumpkin"],
    "Color": ["Orange", "Green", "Green", "Red", "Brown", "White", "Green", "Green", "Green", "Orange"],
    "Calories_per_100g": [41, 34, 23, 18, 77, 40, 15, 25, 81, 26]
}

veges_df = pd.DataFrame(veges_data)
```

```python
veges_df
```

```python
vehicles_data = {
    "Vehicle": ["Car", "Bike", "Bus", "Truck", "Scooter", "SUV", "Van", "Train", "Tram", "Boat"],
    "Top_Speed_kmh": [180, 120, 100, 90, 80, 200, 140, 300, 70, 55],
    "Fuel_Type": ["Petrol", "Petrol", "Diesel", "Diesel", "Petrol", "Diesel", "Petrol", "Electric", "Electric", "Diesel"]
}

vehicles_df = pd.DataFrame(vehicles_data)
```

```python
vehicles_df
```

```python
# concat is mergin two datasetds - non broadcasting

pd.concat([fruits_df, veges_df])
```

```python
# column wise 
pd.concat([fruits_df, veges_df], axis = 1)
```

```python
# Column vise merge with you 4 data sets 
pd.concat([fruits_df, veges_df, vehicles_df], axis = 1)
```

```python
# Merge will merge on column , in ur case there isnt any common column 
pd.merge(fruits_df, veges_df)
```
