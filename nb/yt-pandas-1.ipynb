```python
# Reading Data 

def impo1():

    # Series data 
    s1 = pd.Series([10,20,30,40])
    s1.name = "S1Data"

    # Convert a np arange into series 
    ss1 = np.random.randint(10,50, 10)
    ss1s = pd.Series(ss1)

    # Series data with hetrogenous data 
    s2 = pd.Series([10,20,30,40, 'bubs'])

    

    # Printing 
    rprint(
f"""
s1 Homeogenous series = \n{s1} \n

s1 Hetrogenous Data = \n{s2} \n

---

ss1s = Series from Numpy = 
{ss1s}
Check index =
{ss1s.index}

---

S1 Check data type = {s1.dtype} \n
S1 Check values = {s1.values} \n
S1 Check Index = {s1.index}
S1 DataTable Name = {s1.name}

--- Indexing Data Access ---
Positional Access by index 0 = {s1[0]}
Postional Acess across range = \n {s1[0:4]}
iloc fetch

""")

impo1()
```

```python
# Indexing access 

def ind1():

    # Setting up array and series with mpy 
    is1 = np.random.randint(0,10, 10)
    is1d = pd.Series(is1)

    # Adding Index 
    is1d_index = [
    "Apple", "Banana", "Mango", "Orange", "Pineapple",
    "Grapes", "Strawberry", "Watermelon", "Blueberry", "Papaya"] 

    is1d.index = is1d_index

    # Printing
    rprint(
f"""
is1 = \n {is1} \n

is1d = \n {is1d} \n

--- Index based Access -

Location 0 = {is1d[0]}
Range = \n {is1d[0:5]}

--- Iloc access ---
At position 9 = {is1d.iloc[9]}
Multiple Positions = \n{is1d.iloc[[3,1,5]]}

--- Loc --- 
Calling my label = \n {is1d.loc["Apple": "Grapes"]} \n
Multi Element Access = \n {is1d.loc[["Apple", "Grapes"]]} \n
""")

ind1()
```

```python
# Dictionary Acess 

def dik1():

    boxa("Dictionary and Pandas")

    # Dictionary 
    fr_ca = {
    "Apple": 52.0,
    "Banana": 89.0,
    "Mango": 60.0,
    "Orange": 47.0,
    "Pineapple": 50.0,
    "Grapes": 69.0,
    "Strawberry": 33.0,
    "Watermelon": 30.0,
    "Blueberry": 57.0,
    "Papaya": 43.0
    }

    # Converting the dictionary to series 
    fr_ca_s = pd.Series(fr_ca)
    fr_ca_s.name = "Kalories" 

    # --- Printing ---
    
    pprint(fr_ca_s)
    rprint(
f"""
Fruits and Caliries Pandas Series = \n {fr_ca_s} \n

--- Conditional Acess --- 

Calories>50 =
{fr_ca_s > 50}

Actual Data with condition >50 =
{fr_ca_s[fr_ca_s > 50]}

--- Logical Operators series --- 

Find Fruits Caloreis > 40 & < 80 = 
{(fr_ca_s > 40) & (fr_ca_s < 80)}

Actual data of Above Condition
{fr_ca_s[(fr_ca_s >= 40) & (fr_ca_s <= 80)]}

Or Operation - Yeild all values because of ur range
{fr_ca_s[(fr_ca_s > 60) | (fr_ca_s < 80)]}

Not Operation - yeild all value not greater than 50 (meaning less than)
{fr_ca_s[~(fr_ca_s > 50)]}


 
""")

dik1()
```

```python
# Series Modification

def ser():

    # Defining Series to be modified
    vehicle_speeds = {
    "Car": 120.5,
    "Motorcycle": 98.3,
    "Bicycle": 25.7,
    "Truck": 80.0,
    "Bus": 65.4
    }

    # Convert to Pandas Series
    vh_sp = pd.Series(vehicle_speeds)
    vh_sp.name = "karz"

    # Modify element of series
    # Mod bus speed
    vh_sp["Bus"] = 10.0

    boxa("Series Modifciation")
    pprint(vehicle_speeds)
    rprint(
f"""
-- Data ---
{vh_sp}

--- Mod ---
Make bus = 10.0 \n
{vh_sp["Bus"]}

""")
    
    

ser()
```
