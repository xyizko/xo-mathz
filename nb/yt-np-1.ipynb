```python
# Generating plot with np.linspace values 

# --- Imports ---
import matplotlib.pyplot as plt
import numpy as np

# --- Function ---
def plot_with_linspace():
    # Define axis limits
    xmin = -10
    xmax = 10
    ymin = -10
    ymax = 10

    # Number of points to generate
    points = 20  # High resolution for smooth curve
    
    # Generate x values using linspace
    x = np.linspace(xmin, xmax, points)
    rprint(x)
    
    # Define a function (example: sine wave with linear trend)
    y = 2 * np.sin(x) + 0.3 * x

    # Apply dark background style
    plt.style.use('dark_background')
    fig, ax = plt.subplots(figsize=(10, 6))

    # Set labels and title
    ax.set_xlabel('X-axis', fontsize=12)
    ax.set_ylabel('Y-axis', fontsize=12)
    ax.set_title('Plot Using np.linspace in Dark Mode', fontsize=14, color='white')

    # Set plot dimensions
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)

    # Draw axis lines (centered at 0)
    ax.axhline(0, color='white', linewidth=0.8, alpha=0.7)
    ax.axvline(0, color='white', linewidth=0.8, alpha=0.7)

    # Plot the data
    ax.plot(x, y, color='cyan', linewidth=2, label='y = 2*sin(x) + 0.3x')

    # Add grid
    ax.grid(True, alpha=0.2)

    # Add legend
    ax.legend()

    # Display the plot
    plt.show()

# --- Execution ---
plot_with_linspace()
```

```python
#----------------------
# Making side by side plots to understand difference between 
# np.random.rand - Uniform random distribution 
# np/random.randn - Standard random distribution
#----------------------


import numpy as np
import matplotlib.pyplot as plt

def plot_rand_vs_randn_dark_mode(n_points=500, figsize=(14, 6)):
    """
    Create a single image with two subplots:
    - Left: Scatter plot using np.random.rand (uniform distribution)
    - Right: Scatter plot using np.random.randn (standard normal distribution)
    
    Both in dark mode.
    """
    # Use dark background style
    plt.style.use('dark_background')

    # Generate data
    x_rand = np.random.rand(n_points)
    y_rand = np.random.rand(n_points)
    
    x_randn = np.random.randn(n_points)
    y_randn = np.random.randn(n_points)

    # Create subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)

    # --- Left subplot: rand (uniform) ---
    ax1.scatter(x_rand, y_rand, color='cyan', edgecolors='white', s=40, alpha=0.8)
    ax1.set_title('np.random.rand\n(Uniform Distribution [0, 1))', fontsize=14, color='white')
    ax1.set_xlabel('X values', fontsize=12, color='white')
    ax1.set_ylabel('Y values', fontsize=12, color='white')
    ax1.grid(True, color='gray', linestyle='--', alpha=0.5)

    # --- Right subplot: randn (normal) ---
    ax2.scatter(x_randn, y_randn, color='magenta', edgecolors='white', s=40, alpha=0.8)
    ax2.set_title('np.random.randn\n(Standard Normal Distribution)', fontsize=14, color='white')
    ax2.set_xlabel('X values', fontsize=12, color='white')
    ax2.set_ylabel('Y values', fontsize=12, color='white')
    ax2.grid(True, color='gray', linestyle='--', alpha=0.5)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

# Example usage:
plot_rand_vs_randn_dark_mode(n_points=300)
```

```python
#------------------------
# Plotting 
# np.random.rand = Generate uniform random distribution 
# np.random.randn = Generate normal random distribution 
#------------------------

import numpy as np
import matplotlib.pyplot as plt

def plot_rand_randn_2x2_dark(n_points=100, figsize=(14, 10)):
    """
    Create a 2x2 subplot in dark mode:
    - Top left:  rand scatter
    - Top right: rand line
    - Bottom left:  randn scatter
    - Bottom right: randn line
    """
    # Use dark background style
    plt.style.use('dark_background')

    # Generate data
    x_rand = np.random.rand(n_points)
    y_rand = np.random.rand(n_points)

    x_randn = np.random.randn(n_points)
    y_randn = np.random.randn(n_points)

    # Create figure and subplots
    fig, axs = plt.subplots(2, 2, figsize=figsize)
    fig.suptitle('Comparison: np.random.rand vs np.random.randn\n(Dark Mode)', 
                 fontsize=16, color='white', weight='bold')

    # ----------------------------
    # Top Left: rand - Scatter
    # ----------------------------
    axs[0, 0].scatter(x_rand, y_rand, color='cyan', edgecolors='white', s=40, alpha=0.8)
    axs[0, 0].set_title('rand - Scatter (Uniform)', fontsize=13, color='white')
    axs[0, 0].set_xlabel('X')
    axs[0, 0].set_ylabel('Y')
    axs[0, 0].grid(True, color='gray', linestyle='--', alpha=0.5)

    # ----------------------------
    # Top Right: rand - Line Plot
    # ----------------------------
    axs[0, 1].plot(x_rand, color='lime', label='X Series')
    axs[0, 1].plot(y_rand, color='magenta', label='Y Series', alpha=0.8)
    axs[0, 1].set_title('rand - Line Plot(Uniform)', fontsize=13, color='white')
    axs[0, 1].set_xlabel('Index')
    axs[0, 1].set_ylabel('Value')
    axs[0, 1].legend(fancybox=True, framealpha=0.3)
    axs[0, 1].grid(True, color='gray', linestyle='--', alpha=0.5)

    # ----------------------------
    # Bottom Left: randn - Scatter
    # ----------------------------
    axs[1, 0].scatter(x_randn, y_randn, color='gold', edgecolors='white', s=40, alpha=0.8)
    axs[1, 0].set_title('randn - Scatter (Normal)', fontsize=13, color='white')
    axs[1, 0].set_xlabel('X')
    axs[1, 0].set_ylabel('Y')
    axs[1, 0].grid(True, color='gray', linestyle='--', alpha=0.5)

    # ----------------------------
    # Bottom Right: randn - Line Plot
    # ----------------------------
    axs[1, 1].plot(x_randn, color='lightblue', label='X Series')
    axs[1, 1].plot(y_randn, color='orange', label='Y Series', alpha=0.8)
    axs[1, 1].set_title('randn - Line Plot (Normal)', fontsize=13, color='white')
    axs[1, 1].set_xlabel('Index')
    axs[1, 1].set_ylabel('Value')
    axs[1, 1].legend(fancybox=True, framealpha=0.3)
    axs[1, 1].grid(True, color='gray', linestyle='--', alpha=0.5)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

# Example usage:
plot_rand_randn_2x2_dark(n_points=150)
```

# ArithmeticOperaions

1. Arithmetic operations occur without looping

```python
# Addition of two arrays 

def add_array():

    boxa("Arithmetic Operations")

    a1 = np.array([2,5,8])
    a2 = np.array([2,3,4])
    rprint(f"""
Array a1 = ${a1}
Araay a2 = #{a2}
    """)

    # addition Elementwise operations 
    add_array = a1 + a2
    rprint(f"Addition: ${add_array}")

    # Subtraction
    sub_array = a1 - a2
    rprint(f"Subtraction: ${sub_array}")

    # Multiplication
    mul_array = a1 * a2
    rprint(f"Multiplication: ${sub_array}")

    # Division
    div_array = a1 / a2
    rprint(f"Division: ${div_array} ")

    # Integer Operation - Divison _ Gives whole number
    int_div_array = a1 // a2 
    rprint(f"Integer Division: ${int_div_array} ")

    # Modulus -  Gives Remainder
    modulo_array = a1 % a2 
    rprint(f"Modulo: ${modulo_array}")

    # Exponent -  Power Operation
    pow_array = a1**a2 
    rprint(f"Power: ${pow_array}")
    
    
    

# Execute
add_array()
```

```python
# Make Table - using rich
from rich.console import Console
from rich.table import Table
import numpy as np

def make_tabl1():

    # Create a 3x3 matrix with numbers 1 through 9
    matrix = np.arange(5, 14).reshape(3, 3)
    
    # Create a Rich table
    table = Table(title="3x3 Matrix")
    
    # Add column headers (you can customize these)
    for col_index in range(matrix.shape[1]):
        table.add_column(f"Col {col_index+1}", justify="center")
    
    # Add rows
    for row in matrix:
        table.add_row(*[str(val) for val in row])
    
    # Print table to console
    console = Console()
    console.print(table)

# Execute Function
make_tabl1()
```
