```python
# Check numpy version

def lib_version_checker():
    # Simple script to check library versions

    # List of libraries you want to check
    libraries = ['numpy', 'matplotlib', 'rich', 'IPython']
    
    rprint("Checking library versions...\n")
    
    # Loop through each library and print its version
    for lib_name in libraries:
        try:
            # Import the library dynamically
            lib = __import__(lib_name)
            # Get the version if available
            version = lib.__version__ if hasattr(lib, '__version__') else "Unknown"
            rprint(f"{lib_name:12} : {version}")
        except ImportError:
            rprint(f"{lib_name:12} : Not installed")

lib_version_checker()
```
