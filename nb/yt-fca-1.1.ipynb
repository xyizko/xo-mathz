```python
#--------------------------------------------------------------
# Working Example
#--------------------------------------------------------------
# Complete Example 
import matplotlib.pyplot as plt
import numpy as np

# Apply the built-in dark_background style
plt.style.use('dark_background')

# Generate some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, label='sin(x)')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Plot with dark_background style')

# Add a legend and grid
ax.legend()
ax.grid(True)

# Display the plot
plt.show()
```

```python
# Example with all available color and shepes

import matplotlib.pyplot as plt

def plt1():
    # Define axis limits
    xmin = -10
    xmax = 10
    ymin = -10
    ymax = 10

    # Use dark background style
    plt.style.use('dark_background')
    fig, ax = plt.subplots(figsize=(12, 8))

    # Labels and title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title('Matplotlib: Marker Shapes and Colors', fontsize=14, color='white')

    # Set plot dimensions
    plt.axis([xmin, xmax, ymin, ymax])

    # Draw axis lines (centered at 0)
    plt.plot([xmin, xmax], [0, 0], color='gray', linewidth=1)
    plt.plot([0, 0], [ymin, ymax], color='gray', linewidth=1)

    # --- MARKER TYPES (displayed vertically for each type) ---
    markers = [
        ('.', 'point'),
        ('o', 'circle'),
        ('^', 'triangle up'),
        ('v', 'triangle down'),
        ('<', 'triangle left'),
        ('>', 'triangle right'),
        ('s', 'square'),
        ('p', 'pentagon'),
        ('*', 'star'),
        ('+', 'plus'),
        ('x', 'x'),
        ('D', 'diamond'),
        ('h', 'hexagon'),
        ('1', 'tri_down'),
        ('2', 'tri_up'),
        ('3', 'tri_left'),
        ('4', 'tri_right'),
    ]

    # Basic colors with names
    colors = ['r', 'g', 'b', 'c', 'm', 'y', 'w']
    color_names = ['Red', 'Green', 'Blue', 'Cyan', 'Magenta', 'Yellow', 'White']

    # Place markers on grid
    x_start = -8
    y_start = 6
    x_step = 2
    y_step = 1.5

    for i, (marker, label) in enumerate(markers):
        y = y_start - i * y_step
        # Show marker label
        ax.text(x_start - 1.5, y, label, fontsize=9, color='lightgray', verticalalignment='center')

        # Plot this marker in all colors
        for j, color in enumerate(colors):
            x = x_start + j * x_step
            plt.plot(x, y, marker=marker, color=color, markersize=8)

    # Add color legend at the top
    for j, (color, name) in enumerate(zip(colors, color_names)):
        ax.text(x_start + j * x_step, y_start + 1, name, fontsize=10, color=color,
                horizontalalignment='center', weight='bold')

    # Final touch
    ax.grid(True, alpha=0.2)
    plt.show()

# --- Execution ---
plt1()
```

```python
# Plotting points
# --- Imports --- 

import matplotlib.pyplot as plt

# --- Funtiions ---

def plt1():

    # Defining axis
    xmin = -10
    xmax = +10
    ymin = -10
    ymax = +10
    
    
    # Dark style
    plt.style.use('dark_background')
    fig, ax = plt.subplots()

    # Labels
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title('pt1()')
    ax.grid(True, alpha=0.2)

    # plots dimensions
    plt.axis([xmin, xmax, ymin, ymax])

    # Draw Axis Lines
    plt.plot([xmin,xmax],[0,0])
    plt.plot([0,0],[ymin,ymax])

    # Plotting points
    plt.plot([5],[4], 'ro') # ro = red circle
    plt.plot([2],[4], 'm.') # m. = Magenta Dot
    plt.plot([-3],[4], 'bo') # bo = Bleu Circle
    plt.plot([-3],[2], 'g^') # m. = Green Triangle
        

    # Show plot function
    plt.show()    

# --- Execution ---
plt1()
```

```python
# --- Imports ---
import matplotlib.pyplot as plt

# --- Functions ---
def plt2():
    # Defining axis
    xmin = -10
    xmax = +10
    ymin = -10
    ymax = +10
    
    # Dark style
    plt.style.use('dark_background')
    fig, ax = plt.subplots()

    # Labels
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title('pt2()')  # Fixed function name in title
    ax.grid(True, alpha=0.2)

    # Plot dimensions
    plt.axis([xmin, xmax, ymin, ymax])

    # Draw Axis Lines
    plt.plot([xmin, xmax], [0, 0], linewidth=0.5, color="white")  # X-axis
    plt.plot([0, 0], [ymin, ymax], linewidth=0.5, color="white")  # Y-axis

    # Plotting points using loop
    for x in range(10):
        y = 0.5 * x + 1
        plt.plot([x], [y], "y.", markersize=5)  # âœ… Fixed: use ( ) not [ ]

    # Show plot
    plt.show()

# --- Execution ---
plt2()  # Also fixed: calling plt2(), not plt1()
```
